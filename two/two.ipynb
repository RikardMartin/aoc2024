```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
input = np.array([[7, 6, 4, 2, 1],
[1, 2, 7, 8, 9],
[9, 7, 6, 2, 1],
[1, 3, 2, 4, 5],
[8, 6, 4, 4, 1],
[1, 3, 6, 7, 9]])
```

```python
with open('input.txt') as f:
    lines = f.readlines()

array_of_lines = [np.array(list(map(int, line.split()))) for line in lines]
input = np.array(array_of_lines, dtype=object)
```

```python
input[0]
```

```python
def is_safe(report):
    # report = input[5]
    print('report:', report)

    increasing = None
    unsafe = None

    if len(report) < 2:
        return False
        # print('1')
    else:

        for idx in range(len(report)-1):        
            if unsafe:
                break
            
            if (increasing==None):
                if (report[idx+1] > report[idx]):
                    # print('increasing')
                    increasing = True
                elif (report[idx+1] < report[idx]):
                    # print('decreasing')
                    increasing = False
                else: unsafe = True
            
            if increasing:
                if (report[idx+1] - report[idx]) > 3:
                    unsafe = True
                    # print('> 3')
                if (report[idx+1] <= report[idx]):
                    # print('<=')
                    unsafe = True

            else:
                if (report[idx] - report[idx+1]) > 3:
                    unsafe = True
                    # print('< 3')
                if (report[idx+1] >= report[idx]):
                    unsafe = True
                    # print('>=')
            
        if not unsafe:
            print("safe")
            return True
        else:
            return False

    
```

```python
safe_count = 0

for report in input:
    if is_safe(report):
        safe_count += 1

    else:
        for id in range(len(report)):
            if is_safe(np.delete(report, id)):
                safe_count += 1
                break


print('safe_count', safe_count)
```

```python
7 6 4 2 1: Safe without removing any level.
1 2 7 8 9: Unsafe regardless of which level is removed.
9 7 6 2 1: Unsafe regardless of which level is removed.
1 3 2 4 5: Safe by removing the second level, 3.
8 6 4 4 1: Safe by removing the third level, 4.
1 3 6 7 9: Safe without removing any level.
```
